```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------
# Time Series Dataset
# ---------------------------------------------------

years = [2018, 2019, 2020, 2021, 2022]

sales = [20, 25, 30, 35, 40]

df = pd.DataFrame({
    'Year': years,
    'Sales': sales
})

print("Dataset")
print(df)

# ---------------------------------------------------
# Plot Time Series
# ---------------------------------------------------

plt.figure(figsize=(8,5))

plt.plot(df['Year'], df['Sales'], marker='o')

plt.title("Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Sales")

plt.grid(True)

plt.show()

# ---------------------------------------------------
# 3-Point Moving Average
# ---------------------------------------------------

df['Moving_Average'] = df['Sales'].rolling(window=3).mean()

print("\n3-Point Moving Average")
print(df)

# ---------------------------------------------------
# Exponential Smoothing
# ---------------------------------------------------

alpha = 0.3

df['Exponential_Smoothing'] = df['Sales'].ewm(alpha=alpha).mean()

print("\nExponential Smoothing")
print(df)

# ---------------------------------------------------
# Plot Moving Average
# ---------------------------------------------------

plt.figure(figsize=(8,5))

plt.plot(df['Year'], df['Sales'], marker='o', label='Original Sales')

plt.plot(
    df['Year'],
    df['Moving_Average'],
    marker='o',
    label='Moving Average'
)

plt.title("Moving Average")

plt.xlabel("Year")
plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()

# ---------------------------------------------------
# Plot Exponential Smoothing
# ---------------------------------------------------

plt.figure(figsize=(8,5))

plt.plot(df['Year'], df['Sales'], marker='o', label='Original Sales')

plt.plot(
    df['Year'],
    df['Exponential_Smoothing'],
    marker='o',
    label='Exponential Smoothing'
)

plt.title("Exponential Smoothing")

plt.xlabel("Year")
plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```
