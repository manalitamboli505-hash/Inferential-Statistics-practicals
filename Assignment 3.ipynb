```python
import numpy as np
from scipy import stats
import statsmodels.api as sm

# ---------------- Experiment 1: Correlation ----------------
X = np.array([10, 20, 30, 40, 50])
Y = np.array([12, 25, 33, 45, 58])

r, p_val = stats.pearsonr(X, Y)
print("Experiment 1: Correlation")
print("Correlation coefficient r:", r)
print("p-value:", p_val)
print()

# ---------------- Experiment 2: Simple Linear Regression ----------------
X_const = sm.add_constant(X)  # add intercept
model = sm.OLS(Y, X_const).fit()

intercept, slope = model.params
print("Experiment 2: Regression")
print("Regression Equation: Y =", intercept, "+", slope, "X")
print(model.summary())
print()

# ---------------- Experiment 3: Coefficient of Determination (R²) ----------------
print("Experiment 3: R-squared")
print("R-squared:", model.rsquared)
```
