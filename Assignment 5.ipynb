```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# ---------------------------------------------------
# Lab Experiment 1 : Simple Linear Regression
# ---------------------------------------------------

x = np.array([2,3,5,7,9,10]).reshape(-1,1)
y = np.array([2,5,8,10,12,14])

model = LinearRegression()

model.fit(x, y)

y_pred = model.predict(x)

print("Slope =", model.coef_[0])
print("Intercept =", model.intercept_)

plt.scatter(x, y)
plt.plot(x, y_pred)
plt.title("Simple Linear Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# ---------------------------------------------------
# Lab Experiment 2 : Polynomial Regression Degree 1
# ---------------------------------------------------

coefficients = np.polyfit(x.flatten(), y, 1)

print("\nDegree 1 Polynomial Coefficients")
print(coefficients)

poly1 = np.poly1d(coefficients)

plt.scatter(x, y)
plt.plot(x, poly1(x.flatten()))
plt.title("Degree 1 Polynomial")
plt.show()

# ---------------------------------------------------
# Lab Experiment 3 : Degree 2 and Degree 3 Polynomial
# ---------------------------------------------------

temperature = np.array([0,20,40,60,80,100])
pressure = np.array([0.0002,0.0012,0.006,0.03,0.09,0.27])

# Degree 2
poly2 = np.polyfit(temperature, pressure, 2)

# Degree 3
poly3 = np.polyfit(temperature, pressure, 3)

print("\nDegree 2 Polynomial Coefficients")
print(poly2)

print("\nDegree 3 Polynomial Coefficients")
print(poly3)

curve2 = np.poly1d(poly2)
curve3 = np.poly1d(poly3)

x_range = np.linspace(0,100,100)

plt.scatter(temperature, pressure)

plt.plot(x_range, curve2(x_range), label="Degree 2")
plt.plot(x_range, curve3(x_range), label="Degree 3")

plt.legend()
plt.title("Polynomial Regression")
plt.xlabel("Temperature")
plt.ylabel("Pressure")
plt.show()

# ---------------------------------------------------
# Lab Experiment 4 : R² and RMSE
# ---------------------------------------------------

pressure2 = np.array([0.0002,0.0012,0.0006,0.03,0.09,0.27])

poly3 = np.polyfit(temperature, pressure2, 3)

curve = np.poly1d(poly3)

predicted = curve(temperature)

r2 = r2_score(pressure2, predicted)

rmse = np.sqrt(mean_squared_error(pressure2, predicted))

print("\nR² Score =", r2)
print("RMSE =", rmse)
```
