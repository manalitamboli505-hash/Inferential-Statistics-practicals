```python
import numpy as np
from scipy import stats

# ---------------- LAB Experiment 1: One-sample T-test ----------------
sample = np.random.normal(loc=300, scale=18.5, size=20)
t_stat, p_val = stats.ttest_1samp(sample, 310)
print("One-sample t-test:", t_stat, p_val)

# Z-test version (n > 30)
sample_large = np.random.normal(loc=300, scale=18.5, size=35)
z_stat = (np.mean(sample_large) - 310) / (np.std(sample_large, ddof=1)/np.sqrt(35))
print("Z-test statistic:", z_stat)

# ---------------- LAB Experiment 2: Independent samples T-test ----------------
sample1 = np.random.normal(loc=15, scale=np.sqrt(87.5), size=13)
sample2 = np.random.normal(loc=12, scale=np.sqrt(39), size=11)
t_stat, p_val = stats.ttest_ind(sample1, sample2)
print("Independent t-test:", t_stat, p_val)

# ---------------- LAB Experiment 3: Paired samples T-test ----------------
np.random.seed(0)
data1 = np.random.randn(10)
data2 = data1 + np.random.randn(10)*0.5
t_stat, p_val = stats.ttest_rel(data1, data2)
print("Paired t-test:", t_stat, p_val)

# ---------------- LAB Experiment 4: Chi-squared test ----------------
observed = [15, 12, 18, 10, 15, 30]
expected = [16.7, 16.7, 16.7, 16.6, 16.6, 16.7]
chi2, p_val = stats.chisquare(f_obs=observed, f_exp=expected)
print("Chi-squared test:", chi2, p_val)

# ---------------- LAB Experiment 5: Grades vs Population Mean ----------------
grades = [24, 26, 25, 27, 23, 28, 22, 25, 26, 24]  # example dataset
t_stat, p_val = stats.ttest_1samp(grades, 25)
print("Grades t-test:", t_stat, p_val)
```
